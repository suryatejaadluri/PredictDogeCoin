```python
import twint
import nest_asyncio
import pandas as pd 
import datetime
```

```python

nest_asyncio.apply()

keywords = ['#dogecoin', '#doge', '#dogecoinrise', '#tothemoon', '$doge','dogecoin', 'doge', '#doge', 'doge']

twitter_df = pd.DataFrame()

start_date = datetime.date(2021, 3, 16)
end_date = datetime.date(2021, 5, 15)
delta = datetime.timedelta(days=1)
d = datetime.timedelta(days = 2)

while start_date < end_date:
    current_date = start_date
    for k in keywords:
        t = twint.Config()
        t.Search = k
        t.Since = str(current_date)
        t.Until = str(current_date + d)
        t.Lang = 'en'
        t.Store_csv = True
        t.Pandas = True
        twint.run.Search(t)
        twint.storage.panda.save
        Tweets_df = twint.storage.panda.Tweets_df
        twitter_df = pd.concat([twitter_df, Tweets_df], axis = 0)
    
    start_date += delta

```

```python
# Performing the sentiment analysis on the cleaned twitter data

# Importing required modules

import pandas as pd
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer



twitter_data = pd.read_csv('/Users/EPI Project/march-april.csv')

#Eliminating the duplicate tweets

twitter_data = twitter_data.drop_duplicates(subset = 'id', keep = 'first').reset_index(drop = True)

# Performing the sentiment analysis

# Removing stopwords and tokenizing the tweets. 

clean_tweets = []
tweets = [str(tweet) for tweet in twitter_data.tweet]
stop_words = set(stopwords.words('english'))


for item in tweets:
    
    tokenized_words = word_tokenize(item)
    cleaned_tweet = [w for w in tokenized_words if not w in stop_words]
    cleaned_tweet = []
    
    for w in tokenized_words:
        if w not in stop_words:
            cleaned_tweet.append(w)
    clean_tweets.append(cleaned_tweet)
    
    

# Filtering the tweets once again with the correct keywords

keywords = set(['#dogecoin', '#doge', '#dogecoinrise', '#tothemoon', '$doge', 'dogecoin', 'doge', '#doge', '#dogefather', 'dogefather'])

doge_check = [1 if len(keywords.intersection(tweet)) > 0 else 0 for tweet in clean_tweets]

doge_check = pd.Series(doge_check, name = 'doge_Check')

twitter_data = pd.concat([twitter_data, doge_check], axis = 1)

doge_twitter = twitter_data[twitter_data['doge_Check'] == 1].reset_index(drop = True)


# Tokenizing the dataset again

doge_tweets = [str(tweet) for tweet in doge_twitter.tweet]
stop_words = set(stopwords.words('english'))

doge_clean_tweets = []

for tweet in doge_tweets:
    
    tokenized_sentence = word_tokenize(tweet)
    cleaned_tweet = [w for w in tokenized_sentence if not w in stop_words]
    doge_clean_tweets.append(cleaned_tweet)

# Lemmatizing

lemon = WordNetLemmatizer()


for t in range(len(doge_clean_tweets)):
    res = []
    for w in doge_clean_tweets[t]:
        res.append(lemon.lemmatize(w))
    doge_clean_tweets[t] = res

    
# Stemming

stemmer = PorterStemmer()



for t in range(len(doge_clean_tweets)):
    res = []
    for w in doge_clean_tweets[t]:
        res.append(stemmer.stem(w))
    doge_clean_tweets[t] = res

# Remove less important symbols

unwanted = ['@', '#', '$', '%', '&', '*', ':', ';', '"', '.', ',', '/', '!',
           "'s", 'http', '1', '2', '3', '4', '5', '6', '7', '8', '9', '0']


for t in range(len(doge_clean_tweets)):
    
    doge_clean_tweets[t] = [c for c in doge_clean_tweets[t] if c not in unwanted]

    
# After getting the cleaned data, sentiment scores can be found now

sia = SentimentIntensityAnalyzer()

doge_tweets = [' '.join(t) for t in doge_clean_tweets]

doge_sentiment_values = []


for t in doge_tweets:
    
    doge_sentiment_values.append(sia.polarity_scores(t))


doge_positive, doge_negative = [], []


for item in doge_sentiment_values:
    
    doge_negative.append(item['neg'])
    doge_positive.append(item['pos'])

doge_sent_scores = [doge_positive[i] - doge_negative[i] for i in range(len(doge_negative))]

# Concatenating the sentiment scores to the original dataframe

doge_negative = pd.Series(doge_negative, name = 'Negative')
doge_positive = pd.Series(doge_positive, name = 'Positive')
doge_sent_scores = pd.Series(doge_sent_scores, name = 'Sentiment')
doge_twitter = pd.concat([doge_twitter, doge_sent_scores, doge_positive, doge_negative], axis = 1)
```
